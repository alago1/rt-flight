```python
import smopy #pip install smopy
%matplotlib inline
```

```python
ui_map = smopy.Map((29.6167, -82.3815, 29.6734, -82.3061), z=20)

x, y = ui_map.to_pixels(29.6471, -82.3477)
x1, y1 = ui_map.to_pixels(29.637, -82.332)
x2, y2 = ui_map.to_pixels(29.6519, -82.3654)

ax = ui_map.show_mpl(figsize=(10, 10))
ax.plot(x, y, 'or', ms=25, mew=2, alpha=0.5, color="purple" );
ax.plot(x1, y1, 'or', ms=25, mew=2, alpha=0.5, color="purple" );
ax.plot(x2, y2, 'or', ms=25, mew=2, alpha=0.5, color="purple" );
```
